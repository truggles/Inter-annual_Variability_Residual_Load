```python
def go(ba, base):
    df = pd.read_csv(base+ba)
    df['dt'] = pd.to_datetime(df['date_time'])
    #print(len(df.index))
    df2 = df[ df['dt'].apply(lambda x: x.year == 2018) ]
    #print(len(df2.index))
    #print(np.max(df2['cleaned demand (MW)']))
    idx = df2['cleaned demand (MW)'].idxmax()
    #print(df.loc[idx])
    return df2.loc[idx, 'dt'].month, np.mean(df2['cleaned demand (MW)'])


info = {}
months = []
tot = 0.
others = 0.
for ba in bas:
    month, mean = go(ba, base)
    print(f"{ba},{month},{mean}")
    months.append(month)
    info[ba] = [month, mean]
    if month > 5 and month < 10:
        tot += mean
    else:
        others += mean

print(tot, others)
```
